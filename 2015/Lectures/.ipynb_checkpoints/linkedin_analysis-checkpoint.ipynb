```python
## all imports
from IPython.display import HTML
import numpy as np
import urllib2
import bs4 #this is beautiful soup
import time
import operator
import socket
import cPickle
import re # regular expressions
import csv


from pandas import Series
import pandas as pd
from pandas import DataFrame

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_context("talk")
sns.set_style("white")
```

```python
f_info = 'C:/projetos/cs109_marstival/linkedin_profile_info.csv'
f_pos_summary = 'C:/projetos/cs109_marstival/linkedin_profile_positions.csv'
f_pos_complete = 'C:/projetos/cs109_marstival/linkedin_profile_positions_details.csv'

dtypes=None
converters = None

columns=['rowid','profile_name', 'profile_name2','headline','location','link']
df_profiles = pd.read_csv(f_info, names=columns, delimiter=";",encoding="utf-8-sig")

columns=['rowid','position', 'period']

df_positions_alt = pd.read_csv(f_pos_summary, names=columns, delimiter=";",encoding="utf-8-sig",  converters=converters)

columns=['rowid','period', 'position','start','end','position_id']

def conv2(x):
    if (re.search("\A[0-9]{4}",x) is not None):
        x = 'January ' + x
    return x

converters = {'start':conv2,'end':conv2}

df_positions = pd.read_csv(f_pos_complete, names=columns, delimiter=";", encoding="utf-8-sig", converters=converters)


df_profiles['rowid']=df_profiles.rowid.astype(int)
df_positions['rowid']=df_positions.rowid.astype(int)
df_positions_alt['rowid']=df_positions_alt.rowid.astype(int)

df_positions['start'] = pd.to_datetime(df_positions.start, format='%B %Y', errors='coerce')
df_positions['end'] = pd.to_datetime(df_positions.end, format='%B %Y', errors='coerce')
```

```python
def todate(date):
    print date
    if (date is None) or (str(date).lower() in ["nat", "nan", ""]):
        return None
    
    d = time.strptime(date,"%B %Y")
    return d

def fromdate(d):
    if (d is None) or (str(d).lower() in ["nat", "nan", ""]):
        return None
    s = time.strftime("%B %Y",d)
    return s

def date_isnull(d):
    if (d is None) or (str(d).lower() in ["nat", "nan", ""]):
        return True
    else:
        return False   
```

```python
#df_profiles.head()

num_of_profiles = len(df_profiles)
df_profiles['employee'] = num_of_profiles*False
df_profiles['transition_to'] = num_of_profiles* ""
df_profiles['transition_to_date'] = num_of_profiles * ""
df_profiles['transition_from'] = num_of_profiles* ""
df_profiles['transition_from_date'] = num_of_profiles * ""

df_profiles['transition_to_date'] = pd.to_datetime(df_profiles.transition_to_date, format='%B %Y', errors='coerce')
df_profiles['transition_from_date'] = pd.to_datetime(df_profiles.transition_from_date, format='%B %Y', errors='coerce')


for p in df_profiles.rowid:
    #execution info
    #print "\n begining for ", p
    df = df_positions[df_positions.rowid == p]
    #df = df.sort_values(by=['start'], ascending=False)
    
    df_current = df[(df.period == "currentPositionsDetails") & (df.position.str.contains("HSBC"))]
    employee = bool(len(df_current))
    
    pre_hsbc = post_hsbc = None
    for i, df_allps in df.iterrows():
        
        if (re.search("HSBC",df_allps.position.upper()) is not None):
            pre_hsbc = df_allps
            #print "\n post about to if ", df_allps.position.upper()
            if post_hsbc is not None:
                #print "\n post about to break ", df_allps.position.upper()
                break
                
        elif pre_hsbc is not None:
            post_hsbc = df_allps
            #print "\n pre about to break ", df_allps.position.upper()
            break
        else:
            post_hsbc = df_allps
            #print "\n pre about to coontin ", df_allps.position.upper()
            continue
    
    #print "\n emp ", employee
    #print "\n pre: ", pre_hsbc
    #print "\n post ", post_hsbc
    
    df_profiles.employee[df_profiles.rowid == p] = employee
    if (employee):
        if post_hsbc is not None:
            df_profiles.transition_from[df_profiles.rowid == p] = post_hsbc['position']
            df_profiles.transition_from_date[df_profiles.rowid == p] = post_hsbc['end']
        if pre_hsbc is not None:
            df_profiles.transition_to[df_profiles.rowid == p] = pre_hsbc['position']
            df_profiles.transition_to_date[df_profiles.rowid == p] = pre_hsbc['start']
    else:    
        if post_hsbc is not None:
            df_profiles.transition_to[df_profiles.rowid == p] = post_hsbc['position']
            df_profiles.transition_to_date[df_profiles.rowid == p] = post_hsbc['start']

        if pre_hsbc is not None:
            df_profiles.transition_from[df_profiles.rowid == p] = pre_hsbc['position']
            df_profiles.transition_from_date[df_profiles.rowid == p] = pre_hsbc['end']

    
#df_profiles.head()
```
